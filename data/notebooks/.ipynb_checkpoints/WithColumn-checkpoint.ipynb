```python
juros = spark.read.csv("/user/marcos/data/exercises-data/juros_selic/juros_selic",sep=";", header="true")
```

```python
# 2. Alterar o formato do campo data para “MM/dd/yyyy”
juros.show(5)
```

```python
# Alterando o formado da data para formato americando
data_juros_format_americano = juros.withColumn("data_formatada",unix_timestamp(col("data"),"dd/MM/yyyy"))
data_formatada = data_juros_format_americano.withColumn("data_formatada",from_unixtime("data_formatada","yyyy/MM/dd")).show()
```

```python
# usando uma única linha para fazer todo o processo 
juros_americano = juros.withColumn("data_format_us", from_unixtime(unix_timestamp(col("data"),"dd/MM/yyyy"),"yyyy/MM/dd"))
juros_americano.show()
```

```python
# 3. Com uso da função from_unixtime crie o campo “ano_unix”, com a informação do ano do campo data
# Coletando somente o ano do campo data
juros_ano = juros_americano.withColumn("ano", from_unixtime(unix_timestamp(col("data"),"dd/MM/yyyy"),"yyyy"))
juros_ano.show()
```

```python
# 4. Com uso de substring crie o campo “ano_str”, com a informação do ano do campo data
# coletando o ano usando o substring
juros_substring = juros_ano.withColumn("ano_substring",substring(col("data"),7,4))
juros_substring.show(3)
```

```python
# 5. Com uso da função split crie o campo “ano_str”, com a informação do ano do campo data
# O separados do split é a barra /, e o getItem irá pegar o 3 campo do array que o split cria 
juros_split = juros_substring.withColumn("ano_split",split(col("data"),"/").getItem(2))
juros_split.show(3)
```

```python
# 6. Salvar no hdfs /user/rodrigo/juros_selic_americano no formato CSV, incluindo o cabeçalho
juros_split.write.csv("/user/marcos/juros_selic_formatado",header="true")
```

## Exercícios - With Column parte 2

```python
# imports
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
#!hdfs dfs -ls /user/marcos/data/exercises-data/juros_selic

# 1. Criar um dataframe para ler o arquivo no HDFS /user/<nome/data/juros_selic/juros_selic
juros_selic = spark.read.csv("/user/marcos/data/exercises-data/juros_selic/juros_selic", header="true", sep=";")
```

```python
juros_selic.show(3)
```

### Transformações
Fazendo as transformações dos campos 

```python
# Inserir o campo de ano no dataframe
# mostrando a função split em funcionamento
# juros_ano = juros_selic.withColumn("ano",split(col("data"),"/")).show(3)
juros_ano = juros_selic.withColumn("ano",split(col("data"),"/").getItem(2))
juros_ano.show(3)
```

```python
#  Alterando o campo de valor ou criando um novo para no lugar da virgula ser um ponto (.)
juros_valor_ponto = juros_ano.withColumn("valor_ponto",regexp_replace(col("valor"),"\,","."))
juros_valor_ponto.show(3)

# o campo de valor_ponto continua string tem que fazer um cast para mudar para double ou numeric
juros_valor_ponto.printSchema()
```

```python
# alterar o tipo do valor_ponto para float usando o cast
# pode ser criado outro dataframe porém foi feito em uma única linha
juros_valor_ponto = juros_ano.withColumn("valor_ponto",regexp_replace(col("valor"),"\,",".").cast(FloatType()))
juros_valor_ponto.show(3)
juros_valor_ponto.printSchema()
```

```python
# fazendo GoupBy e Agregações

juros_relatorio = juros_valor_ponto.groupBy("ano").agg(count("ano").alias("meses") ).show()
```

```python
# Incluindo o valor medio e formantando o campo de numeros para duas casas decimais
juros_relatorio = juros_valor_ponto.groupBy("ano").agg(count("ano").alias("meses"),format_number(avg("valor_ponto"),2).alias("valor_medio") ).show()

```

```python
# incluindo o valor minimo e maximo e ordenação
juros_relatorio = juros_valor_ponto.groupBy("ano").agg(count("ano").alias("meses"),
                                                       format_number(avg("valor_ponto"),2).alias("valor_medio"), 
                                                       min("valor_ponto").alias("valor_minimo"),
                                                       max("valor_ponto").alias("valor_maximo")).sort(desc("ano"))
```

```python
juros_relatorio.show()
```

```python
# 3. Salvar no hdfs:///user/<nome>/relatorioAnual com compressão zlib e formato orc
# Salvando o dataframe no hdfs

juros_relatorio.write.orc("/user/marcos/relatorio_anual",compression="zlib")
```
